```python
import numpy as np
import pandas as pd
import nltk
import helper
```

```python
tweets_raw = pd.read_csv('../../../data/twitter/twitter_2L.csv', index_col=0)
# tweets_raw = pd.concat([tweets_raw[:1000],tweets_raw[-1000:]], axis=0)
tweets_raw.columns = ['target','id','date','query','user','text']
tweets_raw['target'] = tweets_raw['target'].map({0:0,4:1})
tweets_raw
```

```python
tweets_raw['clean_text'] = tweets_raw.apply(lambda x: helper.process_tweet(x['text']), axis=1)
tweets_raw
```

```python
tweet_words = tweets_raw[['id','clean_text','target']].explode('clean_text').dropna(axis=0, subset=['clean_text'])
tweet_words_count = tweet_words.groupby(['clean_text','target']).count().reset_index().rename(columns={'id':'count','clean_text':'word'})
tweet_words_count
# tweet_words_count[tweet_words_count['clean_text']=='updat']
```

```python
# negative_words = tweet_words.loc[tweet_words['target']==0]
# neg_words = negative_words.groupby('id').sum()['count'].reset_index()
# neg_words
```

```python
tweet_words = tweet_words.merge(tweet_words_count[tweet_words_count['target']==1][['word','count']], how='left', left_on='clean_text', right_on='word').rename(columns={'count':'pos_count'})
tweet_words = tweet_words.merge(tweet_words_count[tweet_words_count['target']==0][['word','count']], how='left', left_on='clean_text', right_on='word').rename(columns={'count':'neg_count'})
```

```python
tweet_features = tweet_words.groupby('id').sum()[['pos_count','neg_count']].reset_index()
tweet_features
```

```python
tweets_raw=tweets_raw.merge(tweet_features, how='left', left_on='id', right_on='id').fillna(0)
tweets_raw
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
X,Y = tweets_raw[['pos_count','neg_count']],tweets_raw[['target']]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
svc_clf = SVC() 
log_clf = LogisticRegression()
nb_clf = MultinomialNB()
rf_clf = RandomForestClassifier(n_estimators=20)
xgb_clf = XGBClassifier(n_estimators=20)

svc_clf.fit(X_train, Y_train)
log_clf.fit(X_train, Y_train)
nb_clf.fit(X_train, Y_train)
rf_clf.fit(X_train, Y_train)
xgb_clf.fit(X_train,Y_train)
```

```python
def results(name, model):
    '''
    Custom function to check model performance on validation set
    '''
    preds = model.predict(X_test)

    print(name + " score: %.3f" % model.score(X_test, Y_test))
    print(classification_report(Y_test, preds))
    labels = ['Male', 'Female']

    conf_matrix = confusion_matrix(Y_test, preds)
    plt.figure(figsize= (10,6))
    sns.heatmap(conf_matrix, xticklabels=labels, yticklabels=labels, annot=True, fmt="d", cmap='Blues')
    plt.title("Confusion Matrix for " + name)
    plt.ylabel('True Class')
    plt.xlabel('Predicted Class')
```

```python
results("SVC" , svc_clf)
results("Logistic Regression" , log_clf)
results("Naive Bayes" , nb_clf)
results("Random Forest" , rf_clf)
results("XGBoost", xgb_clf)
```

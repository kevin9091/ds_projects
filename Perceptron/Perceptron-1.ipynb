```python
import numpy as np
```

```python
class Perceptron:
    """Implements perceptron network. This object holds only parameters W, and hyperparmaeters lr and epoch"""
    def __init__(self, input_size, lr=1, epochs=100):
        self.W = np.zeros(input_size+1)
        self.epochs = epochs
        self.lr = lr
    
    def activation_fn(self, x):
        return(x>=0).astype(np.float32)
    
    def predict(self, x):
        z = self.W.T.dot(x)
        print('z:',z)
        a = self.activation_fn(z)
        print('a:',a)
        return(a)
        
    def fit(self, X, d):
        for _ in range(self.epochs):
            for i in range(d.shape[0]):
                print('epoch:',_)
                x = np.insert(X[i],0,1)
                print('x:',x)
                y = self.predict(x)
                print('y:',y)
                print('d:',d[i])
                e = d[i] - y
                print('e:',e)
                self.W = self.W + self.lr*e*x
                print('weights:',self.W)
```

```python
if __name__ == '__main__':
    X = np.array([
        [0, 0],
        [0, 1],
        [1, 0],
        [1, 1]
    ])
    d = np.array([0, 0, 0, 1])

    perceptron = Perceptron(input_size=2,epochs=12)
    perceptron.fit(X,d)
```

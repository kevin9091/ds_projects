```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = pd.read_csv('../../data/netflix_titles.csv')
ratings = pd.read_csv('../../data/IMDB_ratings.csv')
ratings.head()
```

```python
data['IMDB_rating'] = ratings['IMDB_rating'].str.split().str[0].str.replace('Not','0.0')
data['Number of reviews'] = ratings['IMDB_rating'].str.split().str[3].str.replace(',','')
data['IMDB_rating'] = data['IMDB_rating'].astype('float')
data['Number of reviews'] = data['Number of reviews'].dropna().astype('int')
# data = data.dropna()
data.head()
```

### Actors Analysis

```python
cast_data = data[['show_id','type','cast','date_added','release_year','rating','duration','description','IMDB_rating','Number of reviews']]
# cast_data.head()
cast_data.shape
```

```python
cast_data['cast'] = cast_data['cast'].str.split(', ')
```

```python
cast_data = cast_data.explode('cast').reset_index()
cast_data.head()
# cast_data.shape
```

```python
### Most common movie actors
# cast_data[cast_data['type']=='Movie'].groupby('cast').count()['index'].sort_values(ascending=False)[:20]
cast_data['movie_count'] = cast_data.groupby('cast')['index'].transform('count')
cast_data
```

```python
### Most highly rated movie actors
cast_data[cast_data['type']=='Movie'].groupby('cast').mean()['IMDB_rating'].sort_values(ascending=False)[:20]
```
